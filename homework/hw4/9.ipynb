```python
import numpy as np
from scipy import stats
from Bio import SeqIO
```

```python
P = np.array([
    [0.224, 0.248, 0.390, 0.138],
    [0.251, 0.360, 0.101, 0.288],
    [0.197, 0.300, 0.341, 0.162],
    [0.136, 0.277, 0.396, 0.190]
])
P = P / P.sum(axis=1)[:, np.newaxis]
```

```python
record = next(SeqIO.parse("Homo_sapiens.GRCh38.dna.chromosome.1.fa", "fasta"))
seq = str(record.seq[10000:20000]).upper()

nucs = ['A', 'C', 'G', 'T']
row_sums = np.array([seq[:-1].count(n) for n in nucs])
total = row_sums.sum()

observed = P * row_sums[:, None]
expected = np.outer(row_sums, observed.sum(axis=0)) / total
chi2 = np.sum((observed - expected)**2 / (expected + 1e-9))
p_val = stats.chi2.sf(chi2, df=9)

print(f"\nхи: {chi2}\nпи вэлюе: {p_val}")
print(("зависимы" if p_val < 0.05 else "независимы"))
```
