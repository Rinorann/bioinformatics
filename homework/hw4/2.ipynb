```python
from Bio import motifs, SeqIO
from Bio.Seq import Seq
```

```python
sites = [ "GAGGTAAAC" , "TCCGTAAGC" , "CAGGTTGGA" ,
 "ACAGTCAGC" , "TAGGTCAGC" , "CAGGTCAGC" ,
 "CAGGTCGAT" , "CAGGTCAGC" , "CAGGTCAGC" ,
 "CAGGTTGGC" ]
```

```python
instances = [Seq(s) for s in sites]
```

```python
m = motifs.create(instances)
```

```python
import urllib.request
import gzip
import shutil
import os

# URL файла
url = "ftp://ftp.ensembl.org/pub/current_fasta/homo_sapiens/dna/Homo_sapiens.GRCh38.dna.chromosome.1.fa.gz"
filename = "Homo_sapiens.GRCh38.dna.chromosome.1.fa.gz"

# Скачивание
print("Скачивание файла... (это займет несколько минут)")
urllib.request.urlretrieve(url, filename)
print("✅ Скачивание завершено!")

# Распаковка
print("Распаковка файла...")
with gzip.open(filename, 'rb') as f_in:
    with open("Homo_sapiens.GRCh38.dna.chromosome.1.fa", 'wb') as f_out:
        shutil.copyfileobj(f_in, f_out)
print("✅ Распаковка завершена!")

# Проверка
if os.path.exists("Homo_sapiens.GRCh38.dna.chromosome.1.fa"):
    file_size = os.path.getsize("Homo_sapiens.GRCh38.dna.chromosome.1.fa") / 1e6
    print(f"📊 Размер файла: {file_size:.2f} MB")
```

```python
# Если скачали с Ensembl (имя файла длинное)
with open("Homo_sapiens.GRCh38.dna.chromosome.1.fa", "r") as handle:
    for record in SeqIO.parse(handle, "fasta"):
        print(f"Загружена хромосома: {record.id}")
        print(f"Длина: {len(record.seq)}")
        chromosome_seq = record.seq
        break
```

```python
sample = chromosome_seq[:1000000]
pssm = m.pssm
```

```python
threshold = 5.0
hits = []
for position, score in pssm.search(sample, threshold=threshold):
    hits.append((position, "+", score))
```

```python
pwm_rc = m.reverse_complement().pssm
for position, score in pwm_rc.search(sample, threshold=threshold):
    hits.append((position, "-", score))
```

```python
hits.sort(key=lambda x: x[0])
for pos, strand, score in hits:
    print(f"{pos}\t\t{strand}\t{score:.4f}")
```
